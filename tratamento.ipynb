# Fazendo transformações

Aqui é interessante usar o método apply ao invés do replace direto, assim não gera erro a partir da segunda utilização

```python
# ['CNPJ_CPF'] para string sem caracteres especiais
ceaps_19_21['CNPJ_CPF'] = ceaps_19_21['CNPJ_CPF'].astype(object)
ceaps_19_21['CNPJ_CPF'] = ceaps_19_21['CNPJ_CPF'].apply(lambda x: str(x).replace(".",""))
ceaps_19_21['CNPJ_CPF'] = ceaps_19_21['CNPJ_CPF'].apply(lambda x: str(x).replace("/",""))
ceaps_19_21['CNPJ_CPF'] = ceaps_19_21['CNPJ_CPF'].apply(lambda x: str(x).replace("-",""))


# ['DATA'] para datetime
# Havia um erro na linha 417 que foi tratado
ceaps_19_21['DATA'] = pd.to_datetime(ceaps_19_21['DATA'], format='%d/%m/%Y', errors='coerce')


# ['VALOR_REEMBOLSADO'] para float
ceaps_19_21['VALOR_REEMBOLSADO'] = ceaps_19_21['VALOR_REEMBOLSADO'].apply(lambda x: str(x).replace(",",".")).astype(float)

# Visualizando
ceaps_19_21.info()
ceaps_19_21.sample(15)
```

```python
# Define uma função para verificar se o número é CPF ou CNPJ
def verificar_tipo_documento(documento):
    if len(documento) == 11:
        return 'CPF'
    elif len(documento) == 14:
        return 'CNPJ'
    else:
        return np.nan


# Aplica a função na coluna 'documento' usando a função where()
ceaps_19_21['TIPO_CNPJ_CPF'] = np.where(ceaps_19_21['CNPJ_CPF'].str.len().isin([11, 14]), 
                                ceaps_19_21['CNPJ_CPF'].apply(verificar_tipo_documento), 
                                np.nan)


# Visualizando
ceaps_19_21.info()
ceaps_19_21.sample(15)
```
